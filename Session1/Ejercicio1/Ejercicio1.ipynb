```python
import math
import numpy as np
import matplotlib.pyplot as plt


x = range(10)

y1 = []
y2 = []
for i in x:
    y1.append(math.exp(-i/10)*np.sin(math.pi*i))    
    y2.append(math.exp(-i/3))


plt.figure(1)
plt.plot(x, y2)
plt.suptitle('primera figura')
plt.figure(2)
plt.plot(x,y1, 'r')
plt.suptitle('segunda figura')
plt.show()
```
